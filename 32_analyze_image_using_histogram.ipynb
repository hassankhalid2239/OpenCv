```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_w=np.ones((500,500,3),np.uint8)*255
blank_b=np.zeros((500,500,3),np.uint8)*255
```

```python
# hist=cv2.calcHist([blank_w],[0],None,[256],[0,256])
hist=cv2.calcHist([blank_b],[0],None,[256],[0,256])

plt.plot(hist)
plt.show()


cv2.imshow("blank_w",blank_w)
cv2.imshow("blank_b",blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread("images\\img6.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hist=cv2.calcHist([img],[0],None,[256],[0,256])

plt.plot(hist)
plt.show()


cv2.imshow("blank_w",blank_w)
cv2.imshow("blank_b",blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread("images\\img6.png")
r,b,g=cv2.split(img)
hist1=cv2.calcHist([r],[0],None,[256],[0,256])
hist2=cv2.calcHist([b],[0],None,[256],[0,256])
hist3=cv2.calcHist([g],[0],None,[256],[0,256])

plt.subplot(3,1,1)
plt.plot(hist1)

plt.subplot(3,1,2)
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)

plt.show()


cv2.imshow("blank_w",blank_w)
cv2.imshow("blank_b",blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
